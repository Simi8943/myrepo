# Data Pre-processing 

To find the missing values

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

To remove the duplicate values

```python
df = df.drop_duplicates()
```

```python
df.info()
```

Creating Derived Columns (Using family_history_with_overweight)

```python
df['Family_History_Binary'] = df['family_history_with_overweight'].map({
    'yes': 1,
    'no': 0
})
```

Classify obesity risk based on family history

```python
df['Family_History_Risk'] = df['family_history_with_overweight'].apply(
    lambda x: 'High Risk' if x == 'yes' else 'Low Risk'
)
```

To display the result

```python
df[['family_history_with_overweight',
    'Family_History_Binary',
    'Family_History_Risk',
    ]].head()
```

To filter the data using 'family_history_with_overweight'] == 'yes'

```python
family_history_yes = df[df['family_history_with_overweight'] == 'yes']
family_history_yes.head()
```

Shows the percentage of obese individuals in each family history group

```python
avg_activity = df.groupby(
    'family_history_with_overweight'
)['FAF'].mean()

avg_activity
```

# Visualizations #

Pie Chart â€“ Gender Split

## Key Insights (Summarized from EDA & Visualizations)


- **Weight is the strongest predictor** of obesity category. Individuals with higher weight consistently fall into overweight or obesity groups.
- **Age distribution**: Majority of participants are young adults (18â€“30), but obesity is also present in older age groups.
- **Family history**: Participants with family history of overweight are more likely to be obese.
- **Lifestyle factors**:
  - Higher physical activity (FAF = Frequently/Always) is associated with Normal Weight.
  - Sedentary lifestyle and high screen time (TUE = Always) correlate with obesity.
- **Dietary habits**:
  - Frequent consumption of high-calorie foods (FAVC = yes) and low vegetable intake (FCVC = 1â€“2) are linked to obesity.
- **Transportation mode**:
  - Public transportation is common across all categories.
  - Automobile users appear more frequently in overweight/obese groups.


## Significant Patterns, Correlations, and Anomalies


- **Correlation Heatmap** shows:
  - Strong positive correlation between Weight and Obesity Level.
  - Negative correlation between Physical Activity (FAF) and Obesity.
- **Anomalies**:
  - Some individuals with relatively low weight are classified as obese, possibly due to height or lifestyle factors.
  - A few participants with high weight but classified as Normal Weight suggest potential misclassification or outliers.


## Visualizations (Reference Plots)


- Bar plots show obesity distribution across categories.
- Histograms highlight skewed distributions in weight and age.
- Box plots reveal clear separation of weight ranges by obesity category.
- Scatter plots show clustering of obesity categories by height and weight.
- Heatmaps confirm correlations between lifestyle factors and obesity.
- Interactive Plotly charts allow deeper exploration of age vs weight trends.


## Conclusion
This analysis highlights the multifactorial nature of obesity:
- **Weight and family history** are primary risk factors.
- **Lifestyle choices** (diet, physical activity, alcohol use, screen time) significantly influence outcomes.
- **Socioeconomic proxies** (transportation mode) may indirectly affect obesity risk.

## Recommendations
- Promote **healthy eating habits** (increase vegetable intake, reduce high-calorie foods).
- Encourage **regular physical activity** to reduce obesity risk.
- Target **awareness programs** for individuals with family history of overweight.
- Further analysis could include **predictive modeling** (e.g., logistic regression, decision trees) to classify obesity risk more accurately.
